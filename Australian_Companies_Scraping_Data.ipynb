```python
from bs4 import BeautifulSoup
import requests 
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_Australian_companies'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
soup.find_all('th')
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_tables_titles = [title.text.strip() for title in world_titles]

print(world_tables_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_tables_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_raw_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_raw_data
    
```
